```python
import pandas as pd
import numpy as np
import glob
from tqdm import tqdm
import cv2
import shutil
from sklearn.model_selection import train_test_split
import os
```

```python
# delete all not jpeg files from new data
web_dir = '/media/yuri/A/constr_engi/photos_old/new'
extensions=['png','jpeg','jpg']
for folder,_,files in os.walk(web_dir):
    for file in files:
        if not any([file.lower().endswith(ext) for ext in extensions]):
            os.remove(os.path.join(folder,file))
```

```python
DIR_TRAIN = "/media/yuri/A/constr_engi/train_dataset/train/"
PATH_TRAIN ="/media/yuri/A/constr_engi/train_dataset/train.csv"
web_dir = '/media/yuri/A/constr_engi/photos_old'
classes={'bulldozer':7,'car':1,'dump truck':4,'excavator':2,'mining loader':5,'soil roller':6,'truck crane':0}
df = pd.read_csv(PATH_TRAIN)
df.head()
```

```python
len(df)
```

```python
num = 0
for folder,_,files in os.walk(web_dir):
    for file in files:
        new_file = os.path.join(folder,file)
        new_name = 'web_'+str(num)+'.jpg'
        cl = float(classes[os.path.split(folder)[-1]])
        shutil.copy(new_file,DIR_TRAIN)
        os.rename(os.path.join(DIR_TRAIN,file),os.path.join(DIR_TRAIN,new_name))
        num+=1
        df=df.append(pd.DataFrame(columns=df.columns,data=[[new_name,cl]]))
```

```python
df.to_csv("/media/yuri/A/constr_engi/train_dataset/train1.csv",index=False)
```
